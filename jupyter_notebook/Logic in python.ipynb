# if-else statement
if there is two condition, we generally uses the if-else chain, you don't need to use if else chain when the if you don't need to process to second condition, just use if statement, if the condition is true
the code will execute else it'll not like:

```python
# this is a test where rahul is greater than or equal to 18 or not
rahul = 12
if rahul >= 18:
    ("you can vote here!")

# here the condition fails
# and nothing happens in this program
```

```python
a = 25
b = 30

if b > a:
    print(f"{b} is greater than {a}")
else:
    print(f"{b} is less than or equal to {a}")
```

We assigned two variables with each value 25 and 30, a is assigned with 25, and b is assigned with 30
we than used if statement with else, the logic is if b is greater than a, in plain english, 30 is greater than 25 = True,
print first argument which is `print(f"{b} is greater than {a}")` which is "30 is greater than 25", if we change the value of b, imagine we use 5
than it'll print second argument right after else, it would be "5 is less than or equal to 25".

# if statements
if you only have one condition, using if-else chain is useless
because if first test fails the second test executes or if first test passes the second test fails

an example: if there is an thing on list how can we print it? what we do?:

```python
fruits = ['apple', 'blueberry', 'dragonfruit']
if 'apple' in fruits:
    print("apple is in the list")
else: 
    print()
```

now it'll only prints "apple is in the list" because else condition print is empty nothing gots printed, this method is wrong, we can just use as many if statement as we want like

```python
fruits = ['apple', 'blueberry', 'dragonfruit']
if 'apple' in fruits:
    print("apple is in the list")
if 'dragonfruit' in fruits:
    print("dragonfruit is in list")
```

# Logical Operators
There are five important logical operators

1. And (&), python keyword: and
if both conditions are true than true else false,
|   | p | q | p&q |
|---|---|---|-----|
| 1 | T | T | T   |
| 2 | T | F | F   |
| 3 | F | T | F   |
| 4 | F | F | F   |

2. OR(V) keyword: or
if both conditions are false than false else true,
|   | p | q | pvq |
|---|---|---|-----|
| 1 | T | T | T   |
| 2 | T | F | T   |
| 3 | F | T | T   |
| 4 | F | F | F   |

3) Not(! / ~ ) keyword: ! (not), whenever you see ! in python it just means not, but ! sign is only use for ! not equal to
to use that operator there is a special python keyword: not
|   | p | ~p | 
|---|---|--- |
| 1 | T | F  | 
| 2 | F | T  |

4) Logical Implication (->), keyword in python: there is no keyword, but important logic
|   | p | q | p->q |
|---|---|---|------|
| 1 | T | T | T    |
| 2 | T | F | F    |
| 3 | F | T | T    |
| 4 | F | F | T    |

5) Biconditional (Logical Equivalence or XNOR), keyword in python: there is no keyword, but important logic
|   | p | q | p<->q |
|---|---|---|------|
| 1 | T | T | T    |
| 2 | T | F | F    |
| 3 | F | T | F    |
| 4 | F | F | T    |

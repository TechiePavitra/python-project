# Tuples in Python
Tuples is a collection of items that cannot be modify later or change later, We can create a Tuple just by the list syntax instead of using Square Brackets  `['xyx']` use parentheses `( )` but parentheses doesn't define a tuple we need to actually use `,` comma, even if the element is one like 
`test = (10,)` not `test = (10)` it is just number 10 assigned with `test` variable. 

```python
# Right way to write an Tuple
dimensions = (250, 50) # imagine we have a certain rectangle with certain size which is absolute (fixed).
print(dimensions[0]) # used print function and used same syntax as lists to display
print(dimensions[1]) # used print function and used same syntax as lists to display
```

# Common Mistakes with Tuples
There are two most common issues with tuple, first is we update tuple as like lists and the second issue is forgetting the comma (`,`) in tuple.

```python
# Wrong Way 1 of Defining Tuple
dimensions = (250, 50) # imagine we have a certain rectangle with certain size which is absolute (fixed).
dimensions[0] = 101 
# We are Trying to Update dimensions tuple index 0.
# But it is important to rememeber that tuple means no changes.
# We can modify it by just replacing tuple with list.
# but its not logical to create a tuple than change it to list.
```

## The Second Wrong Way of Defining The Tuple
The most common syntax error or logical error we creates is forgetting the `,` comma in one element tuple, there is a myth comes in our mind that tuples are just parentheses `( )` but to create tuple even in single element we need to use a comma after the element like in below snippet.


```python
# Wrong Way 2 of Defining Tuple
number = (1,)

# Logical Mistake we makes
number = (1) # here we forgot the comma
```

# Looping Through All Values in a 
You can loop over all the values in a tuple using a for loop, just as you did
with a list:

```python
dimensions = (250, 50)
for dimension in dimensions:
    print(dimension)
```

# Writing Over a Tuple
Although you can’t modify a tuple, you can assign a new value to a variable
that represents a tuple. For example, if we wanted to change the dimensions
of this rectangle, we could redefine the entire tuple:

```python
dimensions = (200, 50) # Original Tuple
print("Original Elements:")
for dimension in dimensions:
    print(dimension)

dimensions = (400, 50) # Modified Tuple
print("\nModified Elements:")
for dimension in dimensions:
    print(dimension)
```

The first four lines define the original tuple and print the initial dimensions. We then associate a new tuple with the variable dimensions, and print
the new values. Python doesn’t raise any errors this time, because reassigning a variable is valid.

When compared with lists, tuples are simple data structures. Use them
when you want to store a set of values that should not be changed throughout the life of a program.

# Looping Through a Slice

## What is a Slice?

A slice is a way to pick some items from a list. Instead of getting all items, you pick just the ones you want.

The format is: `list_name[start:stop]`

- `start` = the position where you start (first item to include)
- `stop` = the position where you stop (this item is NOT included)

For example: `players[4:7]` means "give me items at positions 4, 5, and 6"

---

## Your Code

```python
# List of players
players = ['michael', 'charles', 'alex', 'steve', 'suresh', 'mahesh', 'paresh']

# List of winners / three winners
print("Here are our three winners!")

# Looping Through a Slice
for player in players[4:7]: 
    print(player.title())
```

## What Your Code Does

When you run it, you get:

```
Here are our three winners!
Suresh
Mahesh
Paresh
```

---

## Step 1: Create a List of Players

```python
players = ['michael', 'charles', 'alex', 'steve', 'suresh', 'mahesh', 'paresh']
```

You created a list with 7 player names. Each name has a position number:

| Position | 0 | 1 | 2 | 3 | 4 | 5 | 6 |
|----------|---|---|---|---|---|---|---|
| Name | michael | charles | alex | steve | suresh | mahesh | paresh |

---

## Step 2: Print a Message

```python
print("Here are our three winners!")
```

This line just shows a message on the screen. It prints the text as is.

---

## Step 3: Loop Through a Slice

```python
for player in players[4:7]: 
    print(player.title())
```

This is the main part. Let me explain each piece:

- `for` = start a loop that will repeat
- `player` = temporary variable name (it holds one name at a time)
- `in` = loop through each item
- `players[4:7]` = pick items at positions 4, 5, and 6 only
- `print(player.title())` = print the name with first letter capital

---

## Why [4:7]?

The slice `[4:7]` means:
- Start at: position 4 (suresh)
- Stop before: position 7
- So it includes: positions 4, 5, 6

Think of it like a ruler:

```
0     1       2     3      4       5       6      (7 stops here)
michael charles alex steve suresh mahesh paresh
                        Start here     Stop before here
```

---

## Step 4: What Does .title() Do?

The `.title()` function changes the first letter to capital:

- `'suresh'` becomes `'Suresh'`
- `'mahesh'` becomes `'Mahesh'`
- `'paresh'` becomes `'Paresh'`

---

## How the Loop Works

The loop runs 3 times because we have 3 items in the slice:

**Loop 1:**
- Take `players[4]` which is `'suresh'`
- Apply `.title()` which gives `'Suresh'`
- Print it

**Loop 2:**
- Take `players[5]` which is `'mahesh'`
- Apply `.title()` which gives `'Mahesh'`
- Print it

**Loop 3:**
- Take `players[6]` which is `'paresh'`
- Apply `.title()` which gives `'Paresh'`
- Print it

**Loop ends** because position 7 is not included in the slice

## Try Other Slices

### Get the First 3 Players

```python
for player in players[0:3]:
    print(player.title())
```

Output:
```
Michael
Charles
Alex
```

### Get the Last 2 Players

```python
for player in players[-2:]:
    print(player.title())
```

Output:
```
Mahesh
Paresh
```

### Get Every Other Player

```python
for player in players[0:7:2]:
    print(player.title())
```

This uses `[start:stop:step]` format. The `2` means skip every other item.

Output:
```
Michael
Alex
Suresh
Paresh
```

---

## Things to Remember

1. Positions start at 0, not 1
   - First item is at position 0
   - Second item is at position 1
   - And so on...

2. The stop position is NOT included
   - `[4:7]` includes 4, 5, 6 but NOT 7

3. Use good variable names
   - `player` is clear
   - `p` is not clear

4. `.title()` makes text look nice
   - Great for displaying names
   - Capitalizes first letter of each word

5. Comments help explain your code
   - Other people can understand it
   - You remember what it does later

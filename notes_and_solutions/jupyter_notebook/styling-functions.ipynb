# Styling Functions

## Space and Styling
When you are working with functions, use descriptive function names and argument names. Function names should be meaningful and clearly indicate what the function does. Use lowercase names with underscores when spaces are needed. Follow the same requirements for modules as well. Use concise and descriptive module names so that other programmers can easily understand their purpose.

Use comments to clarify what a function does. Immediately after the function definition, add a docstring that explains the function's purpose. This is very useful for other programmers to understand and use your code. Well-documented code is easier to trust and maintain because its behavior is clearly described.

If you specify a default value for a parameter, no spaces should be used on either side of the equals sign.

Create two blank lines between functions, it makes large code easier to read and understand.

```python
names = ["john", "sarah", "raj"]

for name in names:
    names.remove(name)

print(names)
```

# List Mutation During Iteration
In the code above, we expected the output to be [] (an empty list), but "sarah" was never removed. Why?
This happens because of a problem in Python called "List Mutation During Iteration".

When Python iterates over a list, it keeps track of the current position using an index counter starting from 0, 1, 2, 3...

## Step by Step
1. Starting list:
`names = ["john", "sarah", "raj"]`

`           0        1       2`

2. Iteration 1: index `0` -> "john" is removed:
`names = ["sarah", "raj"]`

`            0       1`

Notice the index has updated. "sarah" shifted to index `0`, "raj" moved to index `1`.

3. Iteration 2: Python jumps to index `1` -> "raj" is removed:
`names = ["sarah"]`

`            0   `    

Python skipped index `0` entirely which is where "sarah" now lives.

## Conclusion
Python's index counter kept moving forward, but the list was shrinking at the same time. 
This mismatch caused `"sarah"` to be skipped and never removed.

This is exactly why you should never modify a list inside `a for loop`.

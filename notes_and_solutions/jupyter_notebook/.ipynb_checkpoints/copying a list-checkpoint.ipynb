# How to Create a Copy of List?
1) First Make a list, you can use `.append()`,`.insert()` or any functions.
2) create a new variable where you assigned the copy here i made `inventors`.
3) first enter the variable name than use slice `[:]`, without this you cannot make any changes to new variable.
4) add or remove things from new list or modify it, to test whether it is working properly or not, use print function or debug it.
    

```python
# Create a List
scientists = ['albert einstein', 'marie curie', 'richard feynman', 'niklo tesla']

# Copy of the list (scientists) 
inventors = scientists[:]

# 2 Names were added in this copied list
inventors.append('charles babbage')
inventors.append('alexander graham bell')

# Display list to prove that one copy has been updated but not the original list "scientist"
print(f"The List of scientist is: {scientists}")
print(f"The List of inventors is: {inventors}")

```

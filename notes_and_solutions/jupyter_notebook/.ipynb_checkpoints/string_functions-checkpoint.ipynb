# How To Change Case of a String?
1) To make the first character of every word use `.title()` function
2) To make the first character of first word use `.capitalize()` function
3) To make all letters upper case, use `.upper()` function
4) To make all letters lower case, use `.lower()` function
5) to remove a particular word or letter use `.removeprefix()` function

```python
title_function = "every first letter is now in upper case"
print(title_function.title())

capitalize_function = "the first letter is now in upper case"
print(capitalize_function.capitalize())

upper_case = "everything is now in uppercase"
print(upper_case.upper())

lower_case = "EVERYTHING IS NOW IN LOWERCASE"
print(lower_case.lower())

a = "Hello, World 2026"
a.removeprefix("2026") # removes 2026 from variable value (string)
```

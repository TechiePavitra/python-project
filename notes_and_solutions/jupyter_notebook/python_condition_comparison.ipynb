# Comparing Two Methods for Identifying Favourite Languages in Python

## Introduction
In this document, we will explore a Python code snippet that demonstrates two different methods for identifying and printing Mohit's favourite programming languages. The focus will be on understanding the differences between these methods and their implications for code efficiency and accuracy.

## Key Concepts
The code provided utilizes a dictionary to store information about an individual named Mohit, including his age, favourite programming languages, and city of residence. The primary goal is to iterate through the list of favourite languages and print specific messages based on the presence of certain keywords, namely `"html"` and `"css".`

The two methods presented differ in their approach to checking for these keywords, which leads to different outcomes in the printed results.

## Code Structure
The code is structured as follows:
1. A dictionary named mohit is defined, containing keys for age, favourite languages, and city.
2. Two methods are implemented to iterate through the list of favourite languages and print messages based on conditions.
3. Each method is followed by comments explaining its functionality and effectiveness.

## Code Examples

```python
mohit = {
    "age": 17,
    "favourite_languages": ["python", "javascript", "css", "html", "c", "rust", "myhtmlcsshub"],
    "city": "surat",
}

# First method
print("First method starts here: ")
for language in mohit["favourite_languages"]:
    if "html" in language: 
        print(f"The {language.upper()} is mohit's favourite language")
    elif "css" in language:
        print(f"The {language.upper()} is mohit's favourite language")
    else:
        print(f"The {language.title()} is mohit's favourite language")

# This method checks if the "html" is in the key value of favourite_languages.
# Look, the myhtmlcsshub is also converted to uppercase.
# Because there is html and css word in there.
# So this method is not great to use

print("\nSecond method starts here: ")
# Second method
for language in mohit["favourite_languages"]:
    if language == "html":
        print(f"The {language.upper()} is mohit's favourite language")
    elif language == "css":
        print(f"The {language.upper()} is mohit's favourite language")
    else:
        print(f"The {language.title()} is mohit's favourite language")

# This method is great; it checks if "html" = "html" or not.
# If it's "myhtmlcsshub," which is not equal to html, then it is not being affected!
```

## _Explanation of the Code_

1. **First Method:**
- This method checks if the substring `"html"` or `"css"` is present in each language string.
- As a result, it incorrectly identifies `"myhtmlcsshub"` as a favourite language containing "html" and prints it in uppercase.
- This method is less precise because it relies on substring matching, which can lead to unintended matches.

2. **Second Method**
- This method checks for exact matches with `"html"` and `"css"`.
- It only prints the languages that exactly match these keywords, ensuring that `"myhtmlcsshub"` is not incorrectly identified.
- This approach is more efficient and accurate, as it avoids false positives.

## Conclusion
In summary, the two methods for identifying Mohit's favourite languages illustrate the importance of precision in programming. The first method, while functional, can lead to misleading results due to its reliance on substring matching. In contrast, the second method provides a more reliable approach by checking for exact matches.

### Note for Future Reference
When comparing methods for string matching in programming:
- **Substring Matching:** Can lead to false positives and should be used cautiously.
- **Exact Matching:** Provides clarity and accuracy, making it the preferred choice in scenarios where precision is critical.

By understanding these differences, you can make informed decisions in your coding practices, ensuring that your programs behave as expected.
